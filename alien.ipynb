```python
import matplotlib.pyplot as plt

# Green alien initial coordinates before and while travelling at slow speed
g1 = [0, 15]
g2 = [3, 21]

# Yellow alien initial coordinates before and while travelling at medium speed
y1 = [0,15]
y2 = [5, 25]

# Red alien initial coordinates before and while travelling at fast speed
r1 = [0, 15]
r2 = [7, 29]

# Plots the graphical lines of the aliens
plt.plot(g1, g2, '-g', label= "Green Alien")
plt.plot(y1, y2, '--y', label = "Yellow Alien")
plt.plot(r1, r2, '-r', label ="Red Alien")

# Labelling of both the axes
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Title of the graph being plotted
plt.title("Aliens Travelling at their featured speed")

# Displays labels of the aliens being plotted
plt.legend()

plt.show()
```
